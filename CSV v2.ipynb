```python
#grabbing the file from folder
pd.read_excel(r"C:\Users\Dani\Downloads\python main\test 1\metadata\all-tokens.xlsx")
#assign variable
dataframe = pd.read_excel(r"C:\Users\Dani\Downloads\python main\test 1\metadata\all-tokens.xlsx")
#print(dataframe)

j, i = 1, 0


for i in dataframe.index:
    entry = dataframe.iloc[i]
    
    #file path
    #files_path = (f'./Random Turtles Collection 1/{i}.png')
    #upload file
    file_input = driver.find_element_by_id('media')
    #file_input.send_keys(files_path)

    #image title
    name_input = driver.find_element_by_name('name')
    name_input.send_keys(entry['name'])

    #image description
    description_input = driver.find_element_by_id('description')
    description_input.send_keys("Enjoy a Randomised 90's Turtle!")


    #click on add properties to cause the box to pop up 
    add_buttons = driver.find_element_by_xpath("//i[@value='add']")
    driver.execute_script("arguments[0].click();", add_buttons)
    
    #add category name 1 - Background
    add_cat_txt_xpath = driver.find_element_by_xpath("//input[@placeholder='Character']")
    add_cat_txt_xpath.send_keys("Background")

    # #add cat type 1 - Background type
    add_cat_value_txt_xpath = driver.find_element_by_xpath("//input[@placeholder='Male']")
    add_cat_value_txt_xpath.send_keys(entry['background'])

    #add more categories
    add_cat_but_more = driver.find_element_by_xpath("//button[@class='Blockreact__Block-sc-1xf18x6-0 Buttonreact__StyledButton-sc-glfma3-0 bhqEJb gIDfxn']")
    driver.execute_script("arguments[0].click();", add_cat_but_more)

    #add category name 2 - Body
    add_cat_txt_xpath = driver.find_element_by_xpath("//input[@placeholder='Character'{1}]")
    add_cat_txt_xpath.send_keys("Body")

    # #add cat type 2 - Body type
    add_cat_value_txt_xpath = driver.find_element_by_xpath("//input[@placeholder='Male']")
    add_cat_value_txt_xpath.send_keys(entry['body'])
    
    #add more categories
    add_cat_but_more = driver.find_element_by_xpath("//button[@class='Blockreact__Block-sc-1xf18x6-0 Buttonreact__StyledButton-sc-glfma3-0 bhqEJb gIDfxn']")
    driver.execute_script("arguments[0].click();", add_cat_but_more)

    #add category name 3 - Costume
    add_cat_txt_xpath = driver.find_element_by_xpath("//input[@placeholder='Character']")
    add_cat_txt_xpath.send_keys("Costume")

    # #add cat type 3 - Costume type
    add_cat_value_txt_xpath = driver.find_element("//input[@placeholder='Male']")
    add_cat_value_txt_xpath.send_keys(entry['costume'])

    #save button
    add_cat_but_more = driver.find_element_by_xpath("//button[@class='Blockreact__Block-sc-1xf18x6-0 Buttonreact__StyledButton-sc-glfma3-0 bhqEJb fzwDgL']")
    driver.execute_script("arguments[0].click();", add_cat_but_more)
```
